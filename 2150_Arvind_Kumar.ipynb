```python
# Load the dataset using pandas

import pandas as pd
dataFile = pd.read_csv("campaign.csv")
print(dataFile.head())
```

```python
# Inspect the structure: column names, data types, null values

listOfColumns = dataFile.columns.tolist()

print(f"Column names are : {listOfColumns}")

dataType = dataFile.dtypes

print(f"\nColumns                  Data types:" f"\n{dataType}")

sumOfNull = dataFile.isnull().sum()

print(f"\nMissing values per column is : {sumOfNull}")
```

```python
# Write a pure function to describe basic dataset info (rows, cols, column names)

def describe_dataFile_basic_dataset(dataFile: pd.DataFrame) -> dict:
    return {
        "num_rows": dataFile.shape[0],
        "num_columns": dataFile.shape[1],
        "column_names": dataFile.columns.tolist()
    }

info = describe_dataFile_basic_dataset(dataFile)
print(f"Dataset Info : {info}")
```

```python
def clean_campaign_data(df):
  
    # Convert to list of dictionaries to use functional tools
    records = df.to_dict(orient='records')

    # Use filter() to remove invalid or incomplete rows
    valid_rows = list(filter(lambda row: pd.notnull(row.get('campaign_id')) 
                                         and pd.notnull(row.get('name')) 
                                         and row.get('org_name') not in [None, '', 'UNKNOWN'], 
                              records))

   # Use list comprehensions or map() to normalize or transform column data with strip text, convert cases
    cleaned_rows = [
        {
            **row,
            'name': row['name'].strip().title() if pd.notnull(row['name']) else '',
            'org_name': row['org_name'].strip().upper() if pd.notnull(row['org_name']) else '',
        }
        for row in valid_rows
    ]

    # Return a new cleaned DataFrame
    return pd.DataFrame(cleaned_rows)


cleaned_dataFile = clean_campaign_data(dataFile)
print(cleaned_dataFile.head())
```

```python
cleaned_data = []
for row in dataFile.to_dict('records'):
    if row['name'] and row['campaign_status']:
        row['campaign_status'] = row['campaign_status'].upper()
        cleaned_data.append(row)

print(f"Cleanned data without functional programming : {cleaned_data}")
```

```python
cleaned_data = list(
    map(
        lambda row: {**row, 'campaign_status': row['campaign_status'].upper()},
        filter(
            lambda row: row.get('name') and row.get('campaign_status'),
            dataFile.to_dict('records')
        )
    )
)

cleaned_df = pd.DataFrame(cleaned_data)
print(cleaned_df)
```
